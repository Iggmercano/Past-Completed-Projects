```python
import pandas as pd
from glob import glob #read_all_files_in_folders
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
#names_of_emotions_AND_folders
emotions = ['angry',
            'disgust',
            'fear',
            'happy',
            'neutral',
            'sad',
            'suprise']
```

```python
#get_directory_of_audio_files
image_list = []
for emo_list in emotions:
    path = glob('C:/Users/Ivan Gil Mercano/Desktop/Image Emotion/' + emo_list + '/*.*')
    image_list.append(path)
image_list
```

```python
from feat import Detector
face_model = "retinaface"
landmark_model = "mobilenet"
au_model = "rf"
emotion_model = "resmasknet"
detector = Detector(face_model = face_model, landmark_model = landmark_model, au_model = au_model, emotion_model = emotion_model)
```

```python
images_aus = []#image_action_units
for path in image_list:
    for file in path:
        image_prediction = detector.detect_image(file)
        images_aus.append(image_prediction.aus())
```

```python
Imagepath = sum(len(row) for row in image_list) #find total number of files
images_ausdf = pd.DataFrame()
for index in range(Imagepath):
    df = pd.DataFrame(images_aus[index].aus())
    images_ausdf = pd.concat([images_ausdf,df],ignore_index=True)
images_ausdf
```

```python
emo_label = []
i = 0
for emotion_file in image_list:
    emotion_name = emotions[i]
    for file in emotion_file:
        emo_label.append(emotion_name)
    i += 1

emotionDf= pd.DataFrame(emo_label)
emotionDf.columns = ['Emotion']
image_features = pd.concat([emotionDf,images_ausdf], axis=1)
image_features
```

```python
#transfer clean data from images_features_all to images_features_all_woNaN
df = pd.DataFrame()
image_features_clean = df
image_features_clean = image_features.dropna() # drop rows for NaN values
```

```python
import pandas as pd
import numpy as np
import io
import pydotplus
import graphviz
import matplotlib.pyplot as plt
from io import StringIO
from IPython.display import Image  
from sklearn import metrics, tree
from sklearn.tree import export_graphviz
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
file = 'C:/Users/Ivan Gil Mercano/Desktop/Image Emotion/Image Features.csv'
dataset = pd.read_csv(file, sep= ',', header = 0)
```

```python
#Defining the length and shape of the data set
print("Dataset Length:: ", len(dataset))
print("Dataset Length:: ", dataset.shape)
```

```python
#Displaying the Data set
print("Dataset Length:: ")
dataset.head()
```

# Stratified Sample Split

```python
xvar = dataset.drop('Emotion', axis = 1)
feature_cols = xvar.columns

X = dataset[['AU01','AU02',
         'AU04','AU05',
         'AU06','AU07',
         'AU09','AU10',
         'AU11','AU12',
         'AU14','AU15',
         'AU17','AU20',
         'AU23','AU24',
         'AU25','AU26',
         'AU28','AU43',]]
y = dataset[['Emotion']]
```

```python
Strat_Sampling = StratifiedShuffleSplit(n_splits=5, test_size=0.25, train_size=0.75, random_state=0)
for training_data_index, testing_data_index in Strat_Sampling.split(X, y):
    X_train = dataset.iloc[training_data_index].loc[:, feature_cols]
    X_test = dataset.iloc[testing_data_index][feature_cols]
    y_train = dataset.iloc[training_data_index].loc[:,'Emotion']
    y_test = dataset.loc[testing_data_index]['Emotion']
```

# Decision Tree GINI

```python
DecisionTreeGini = DecisionTreeClassifier(criterion = "gini")
DecisionTreeGini = DecisionTreeGini.fit(X_train,y_train)
```

```python
dot_data = StringIO()
export_graphviz(DecisionTreeGini, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,feature_names = feature_cols, class_names=['Angry',
                'Disgusted',
                'Fearful',
                'Happy',
                'Neutral',
                'Sad',
                'Suprised'])
from pydot import graph_from_dot_data
(graph, ) = graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

```python
pred_test = DecisionTreeGini.predict(X_test)
print("Accuracy Score:", accuracy_score(y_test, pred_test))
print("Precision Score:", precision_score(y_test, pred_test, average='micro', zero_division=1))
print("Recall Score:", recall_score(y_test, pred_test, average='micro', zero_division=1))
print(classification_report(y_test, pred_test))
```

```python
RandomForest_Gini= RandomForestClassifier(criterion = "gini")
RandomForest_Gini = RandomForest_Gini.fit(X_train,y_train)
```
